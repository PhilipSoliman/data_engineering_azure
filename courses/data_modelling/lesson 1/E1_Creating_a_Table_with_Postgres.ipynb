#### Import the library 
*Note:* An error might popup after this command has executed. If it does, read it carefully before ignoring. 

```python
import psycopg2
from src.database import get_pg_connection, get_pg_cursor
```

```python
conn = get_pg_connection()
conn.set_session(autocommit=True)
```

```python
cur = get_pg_cursor(conn)
```

```python
## TO-DO: Finish writing the CREATE TABLE statement with the correct arguments
from src.database import create_pg_table

try:
    create_pg_table(
        conn,
        "song_library",
        {
            "song_title": "VARCHAR",
            "artist_name": "VARCHAR",
            "release_year": "INT",
            "album_name": "VARCHAR",
            "is_single": "BOOLEAN",
        },
    )
except psycopg2.Error as e:
    print("Error: Issue creating table")
    print(e)
```

```python
try:
    cur.execute("SELECT count(*) FROM song_library;")
except psycopg2.Error as e:
    print("Error: Issue selecting count from table")
    print(e)

print(cur.fetchall())
```

### Insert the following two rows in the table
`First Row:  "Across The Universe", "The Beatles", "1970", "Let It Be", "False"`

`Second Row: "Think For Yourself", "The Beatles", "1965", "Rubber Soul", "False"`

```python
## TO-DO: Finish the INSERT INTO statement with the correct arguments
from src.database import insert_pg_rows

rows = [
    ["Across The Universe", "The Beatles", 1970, "Let It Be", False],
    ["Think For Yourself", "The Beatles", 1965, "Rubber Soul", False],
]
insert_pg_rows(
    conn,
    "song_library",
    ["song_title", "artist_name", "release_year", "album_name", "is_single"],
    rows,
)
```

### TO-DO: Validate your data was inserted into the table. 


```python
## TO-DO: Finish the SELECT * Statement
try:
    cur.execute("SELECT * FROM song_library")
except psycopg2.Error as e:
    print("Error: select *")
    print(e)

row = cur.fetchone()
while row:
    print(row)
    row = cur.fetchone()
```

### And finally close your cursor and connection. 

```python
from src.database import drop_pg_table

cur.close()
drop_pg_table(conn, "song_library")
conn.close()
```
